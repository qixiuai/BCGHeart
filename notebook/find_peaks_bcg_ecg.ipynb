# bcg findpeaks

```python
import pdb

def _parse_integer(bytes):
    int_str = bytes.decode('ASCII')
    integers = []
    #try:
    #integers = list(map(int, int_str.split(',')))
    for s in int_str.split(','):
        try:
            integer = int(s)
        except:
            print(bytes)
            continue
        integers.append(integer)
    #except:
        #raise ValueError("")
    return integers
    
def parse_bcg(bytes):
    data = []
    packet = []
    start_index = 0
    end_index = 0
    for ind, byte in enumerate(bytes):
        if byte == ord('#'):
            if end_index - start_index > 0:
                #print(start_index, end_index)
                integers = _parse_integer(bytes[start_index:end_index])
                data.extend(integers)
            start_index = ind + 1
            end_index = ind + 1
        if byte == ord('\r'):
            end_index = ind
    if end_index - start_index > 0:
        integers = _parse_integer(bytes[start_index:end_index])
        data.extend(integers)
    return data
```

```python
f = open("/home/guo/physio/BCG_data/bcg_ecg_data/bcg_ecg_data(500HZ)/wsx_bcg_wave(500HZ).txt", 'rb')
data = f.read()
f.close()
bcg = parse_bcg(data)
```

```python
bcg_auto_peaks = autopeaks.AutoPeaks(thres=0.75,min_dist=300,fs=500)
```

```python
list(map(bcg_auto_peaks.findpeaks, bcg));
```

```python
bcg_peak_indices = bcg_auto_peaks.peak_indexes
bcg_peak_values = bcg_auto_peaks.peak_values
```

```python
plot([go.Scatter(y=ecg),go.Scatter(y=bcg), go.Scatter(x=bcg_peak_indices, y=bcg_peak_values, mode="markers")])
```
