```python
import numpy as np
from plotly.offline import plot
import plotly.graph_objs as go
```

```python
energy = np.loadtxt("text")
energy = np.asarray(energy)
```

```python
peak_indices = np.loadtxt("resp_indices.txt")
```

```python
plot([go.Scatter(y=energy), go.Scatter(x=peak_indices, y=energy[peak_indices.tolist()], mode="markers")])
```
